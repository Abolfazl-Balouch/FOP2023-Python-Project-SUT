```python
#type your code here...
df_cleaned = df.dropna()
df = df_cleaned
print(df)
```

<div dir = "rtl" style="font-face:IranSansX;direction:rtl;line-height:250%;font-size:small;">
	<font face="IranSansX" size=3>

```python
#type your code here...
data_types = df.dtypes
print("نوع داده هر ستون:\n", data_types)
```

<div dir = "rtl" style="font-face:IranSansX;direction:rtl;line-height:250%;font-size:small;">
	<font face="IranSansX" size=3>
جنس مقادیر ستون‌های  Area و Price  و Price(USD) را از رشته به مقدار عددی تغییر دهید و دیتافریم جدید را جایگزین قبلی کنید.    </font>

```python
#type your code here...
df['Area'] = pd.to_numeric(df['Area'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Price(USD)'] = pd.to_numeric(df['Price(USD)'], errors='coerce')
print(df)
```

<div dir = "rtl" style="font-face:IranSansX;direction:rtl;line-height:250%;font-size:small;">
	<font face="IranSansX" size=3>
        ستون‌هایی که جنس متغیر آن‌ها از نوع Boolean است را به مقادیر عددی تغییر دهید و دیتافریم جدید را جایگزین قبلی کنید. (به ازای True مقدار 1 و به ازای False مقدار 0 جایگزین شود.)
    </font>
</div>

```python
#type your code here...
df.replace({True: 1, False: 0}, inplace=True)
```

<div dir = "rtl" style="font-face:IranSansX;direction:rtl;line-height:250%;font-size:small;">

<div dir = "rtl" style="font-face:IranSansX;direction:rtl;line-height:250%;font-size:small;">
	<font face="IranSansX" size=3>
        یک دیکشنری از نام ستون‌ها تشکیل داده و نام هر ستون را به عنوان key و نام هر ستون به صورت Pascal case را به عنوان value آن در نظر بگیرید.(این 
        دیکشنری در مراحل بعدی مورد استفاده قرار می‌گیرد.)
        <br>
        </font>
        <font face="IranSansX" size=2>
        <a href="https://stackoverflow.com/questions/17326185/what-are-the-different-kinds-of-cases">می‌توانید از این لینک برای آشنایی با Case Styleها استفاده کنید.</a>

```python
#type your code here...
column_dict = {'Area':'Area', 'Room':'Room', 'Parking':'Parking', 'Warehouse':'Warehouse', 'Elevator':'Elevator', 'Address':'Address', 'Price':'Price', 'Price(USD)':'Price(USD)' }
```

<div dir = "rtl" style="font-face:IranSansX;direction:rtl;line-height:250%;font-size:small;">
	<font face="IranSansX" size=3>
        نام ستون‌هایی که طول آن‌ها از 8 کارکتر کمتر است را خروجی دهید.
    </font>
</div>

```python
#type your code here...
selected_columns = [column for column in df.columns if len(column) < 8]
print("نام ستون‌هایی که طول آن‌ها کمتر از 8 کاراکتر است:", selected_columns)
```

<div dir = "rtl" style="font-face:IranSansX;direction:rtl;line-height:250%;font-size:small;">
	<font face="IranSansX" size=3>
نام تمامی ستون‌ها را به صورت یک رشته‌ که نام هر ستون با کارکتر '-' از دیگری جدا شده، خروجی دهید.    </font>
</div>

```python
#type your code here...
columns_string = '-'.join(df.columns)
print("نام تمامی ستون‌ها با کاراکتر '-' جدا شده:", columns_string)
```

<div dir = "rtl" style="font-face:IranSansX;direction:rtl;line-height:250%;font-size:small;">
	<font face="IranSansX" size=3>
با استفاده از دیکشنری ایجاد شده در مراحل قبل، نام ستون‌های دیتافریم را به صورت Pascal case تغییر دهید و در df ذخیره کنید.
    </font>
</div>

```python
#type your code here...
df.rename(columns=column_dict, inplace=True)
```

نمودار توزیع فراوانی (Frequency Distribution Plot) یکی از ابزارهای مهم در تحلیل و تجزیه و تحلیل داده‌هاست. این نمودار به شما کمک می‌کند تا توزیع و فراوانی ارقام یا مقادیر مختلف در یک مجموعه داده را به وضوح مشاهده کنید. 

کاربردهای اصلی نمودار توزیع فراوانی عبارتند از:

1. **نمایش توزیع داده:** این نمودار به شما این امکان را می‌دهد که فهمیده توزیع داده‌ها چگونه است. آیا داده‌ها به صورت یکنواخت توزیع شده‌اند یا تمرکز خاصی دارند؟

2. **شناسایی الگوها و میانگین:** با نمایش توزیع فراوانی، می‌توانید الگوها و میانگین‌های مهم در داده‌ها را شناسایی کنید. این نمودار به شما امکان می‌دهد تا ببینید که آیا مقادیر داده‌ها به سمت یک سو یا دیگر متمایل هستند.

3. **شناسایی انحرافات و پرتکرارت:** می‌توانید از این نمودار برای شناسایی داده‌های پرتکرار (Outliers) یا انحرافات در داده‌ها استفاده کنید. در نمودار، افترا به خطوط یا نقاطی که از الگوی کلی پرتکرار بیرون می‌روند، می‌تواند به شما کمک کند تا داده‌های غیرعادی را شناسایی کنید.

4. **ارائه مقایسه بین دو یا چند گروه:** اگر دو یا چند گروه مختلف دارید، می‌توانید از نمودار توزیع فراوانی برای مقایسه توزیع داده‌های آن‌ها استفاده کنید. این امکان را به شما می‌دهد که تفاوت‌ها و الگوهای گروه‌ها را به وضوح مشاهده کنید.

کلیتاً، نمودار توزیع فراوانی یک ابزار مفید برای درک و تحلیل دقیق‌تر داده‌هاست و می‌تواند در فرایند تصمیم‌گیری و بررسی داده‌ها به شما کمک کند.

Heatmap یک نوع نمودار است که اطلاعات دو بعدی را با استفاده از رنگ نشان می‌دهد. این نمودار اغلب برای نشان دادن روابط و الگوهای مختلف در داده‌های یک ماتریس به کار می‌رود. کاربردهای Heatmap به شرح زیر می‌باشند:

1. **نشان دادن همبستگی:**
   در داده‌های عددی، Heatmap به عنوان یک ابزار مفید برای نشان دادن میزان همبستگی بین متغیرها استفاده می‌شود. مقادیر همبستگی با استفاده از رنگها نشان داده می‌شوند؛ برای مثال، مقادیر مثبت می‌توانند با یک رنگ (معمولاً رنگ گرم) و مقادیر منفی با رنگ دیگر (معمولاً رنگ سرد) نشان داده شوند.

2. **تحلیل ماتریس داده:**
   Heatmap برای نشان دادن الگوهای موجود در ماتریس‌های داده مورد استفاده قرار می‌گیرد. این الگوها می‌توانند اطلاعاتی ارائه دهند که با دقت مشاهده نمی‌شوند.

3. **نمایش توزیع فراوانی:**
   Heatmap می‌تواند برای نمایش توزیع فراوانی متغیرها در داده‌ها مورد استفاده قرار گیرد. به عنوان مثال، می‌توان از Heatmap برای نمایش توزیع احتمالاتی یک متغیر یا توزیع فراوانی دو متغیر به صورت همزمان استفاده کرد.

4. **مدیریت و ارتباط با داده:**
   Heatmap به تحلیل سریع و ارتباط دادن متغیرها با یکدیگر کمک می‌کند. با استفاده از رنگ‌ها و شدت‌ها، اطلاعات به شکل گرافیکی و قابل فهمی نمایش داده می‌شوند.

کلیدی‌ترین ویژگی Heatmap این است که به صورت سریع و با یک نگاه، اطلاعات پیچیده‌ای را که در داده‌های دیگر نمی‌توان به سرعت درک کرد، ارائه می‌دهد. این ابزار به تحلیل و فهم بهتر داده‌ها کمک کرده و تصمیم‌گیری را تسهیل می‌کند.

نمودار Pie Chart یک نمودار دایره‌ای است که برای نمایش توزیع درصدی یک مجموعه از داده‌ها بر اساس بخش‌ها یا گروه‌ها استفاده می‌شود. این نمودار به صورت دایره‌ای تقسیم می‌شود، به طوری که هر بخش از دایره نمایانگر یک دسته یا گروه از داده‌ها است و مساحت هر بخش نمایانگر درصد نسبی آن گروه از کل داده‌ها می‌شود.

کاربردهای Pie Chart عبارتند از:

1. **نمایش توزیع درصدی**: برای نمایش توزیع درصدی مختلف گروه‌ها یا دسته‌ها از یک مجموعه داده.

2. **نمایش قسمت‌ها از یک کل**: وقتی که می‌خواهید نشان دهید هر بخش یا دسته چقدر از کل داده را تشکیل می‌دهد.

3. **برقراری توازن یا عدم توازن**: برای نمایش توازن یا عدم توازن بین اجزاء مختلف یک مجموعه.

4. **نمایش تعداد گروه‌ها**: وقتی که تعداد گروه‌ها نسبت به هم مهم است و می‌خواهید تفاوت‌ها را بصورت چشم‌گیر نمایش دهید.

همچنین، مهم است که از Pie Chart با دقت استفاده شود و در صورتی که تعداد گروه‌ها زیاد باشد یا درصدها نسبت به هم بسیار نزدیک باشند، استفاده از نمودارهای دیگری مانند Bar Chart را در نظر گرفت.

نمودارهای Barplot یکی از ابزارهای مهم و محبوب در تجزیه و تحلیل داده‌ها هستند. این نوع نمودار برای نمایش توزیع داده‌ها در یک یا چند متغیر گسسته (دسته‌ای) استفاده می‌شود. در زیر تعدادی از کاربردهای متداول Barplot آورده شده است:

1. **نمایش فراوانی (تعداد) داده‌ها در یک متغیر گسسته:**
   - مثال: تعداد خانه‌ها بر اساس تعداد اتاق‌ها، تعداد ماشین‌ها بر اساس تعداد درب‌ها، و غیره.

2. **نمایش میانگین یک متغیر عددی در هر دسته:**
   - مثال: نمایش میانگین قیمت خانه‌ها بر اساس تعداد اتاق‌ها.

3. **نمایش توزیع دو متغیر گسسته:**
   - مثال: نمایش تعداد ماشین‌ها بر اساس تعداد درب‌ها و رنگ.

4. **نمایش توزیع دو متغیر (یک متغیر گسسته و یک متغیر عددی):**
   - مثال: نمایش میانگین قیمت خانه‌ها بر اساس نوع محله (محله یک یا محله دو).

5. **نمایش مقایسه بین چند دسته:**
   - مثال: نمایش میانگین درآمد در چند حوزه‌ی شهری یا نمایش تعداد دانش‌آموزان در هر پایه تحصیلی.

در کل، Barplot یک ابزار قوی برای درک سریع توزیع داده‌ها در دسته‌هاست و به تحلیل و تبیین اطلاعات بر اساس دسته‌بندی‌های مختلف کمک می‌کند.
